```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
image = cv2.imread("../Input/Images/sample.png")
plt.axis('off')
plt.imshow(image[:,:,::-1])
cv2.imwrite("../Input/Images/sample_big.jpg", image)
```

```python
downscaled = cv2.resize(image,(int(image.shape[0]*0.8),int(image.shape[0]*0.8)),interpolation=cv2.INTER_LANCZOS4)
plt.axis('off')
plt.imshow(downscaled/255.0)
cv2.imwrite("../Input/Images/sample_small.png", downscaled)
print(f"sr factor is {float(image.shape[0]) / float(downscaled.shape[0])}")
```

```python
upscaled_interpolation = cv2.resize(downscaled,(int(downscaled.shape[0]*1.25),int(downscaled.shape[0]*1.25)),interpolation=cv2.INTER_LANCZOS4)
diff = -0.5+(upscaled_interpolation-image)
plt.axis('off')
plt.imshow(diff[:,:,0], cmap="seismic")
plt.colorbar()
plt.figure()
plt.axis('off')
plt.imshow(diff[:,:,1], cmap="seismic")
plt.colorbar()
plt.figure()
plt.axis('off')
plt.imshow(diff[:,:,2], cmap="seismic")
plt.colorbar()
```
